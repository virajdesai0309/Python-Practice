```python
from gekko import GEKKO

# Create a GEKKO model
m = GEKKO(remote=False)

# Define variables x and y
x = m.Var()
y = m.Var()

# Add equations to the model
m.Equation(3*x + 2*y == 1)
m.Equation(x + 2*y == 0)

# Solve the system of equations
m.solve(disp=True)

# Print the solution values for x and y
print(x.VALUE)
print(y.VALUE)
```
