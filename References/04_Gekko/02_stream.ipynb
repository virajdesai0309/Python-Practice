```python
from gekko import GEKKO, chemical

# Create a GEKKO model
m = GEKKO(remote=False)

# Add compound database
c = chemical.Properties(m)
c.compound('7732-18-5')  # Water (CAS)

# Create flowsheet
fs = chemical.Flowsheet(m)

# Inlet stream (water at 1 atm, 298 K, 1 kmol/s)
feed = fs.stream(fixed=True)
feed.P.value = 101325      # Pa
feed.T.value = 298.15      # K
feed.ndot.value = 1.0      # kmol/s
feed.phase = fs.set_phase(feed, 'liquid')

# Pump
pump = fs.pump()
pump.dp = m.Param(2e5)  # Î”P = 200 kPa

# Connect inlet
fs.connect(feed, pump.inlet)

# --- Create explicit outlet stream ---
outlet = fs.stream(fixed=False)   # New stream object
fs.connect(pump.outlet, outlet)   # Connect pump outlet to stream

# Solve steady state
m.solve(disp=False)

# Print results
print("Inlet Stream:")
print(f"  Pressure: {feed.P.value[0]:.2f} Pa")
print(f"  Temperature: {feed.T.value[0]:.2f} K")
print(f"  Flowrate: {feed.ndot.value[0]:.2f} kmol/s")

print("\nAfter Pump:")
print(f"  Pressure: {outlet.P.value[0]:.2f} Pa")
print(f"  Temperature: {outlet.T.value[0]:.2f} K")
print(f"  Flowrate: {outlet.ndot.value[0]:.2f} kmol/s")
```
