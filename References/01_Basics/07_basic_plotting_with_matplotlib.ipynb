1. Introduction

“Matplotlib is a powerful Python library for creating 2D plots. In chemical engineering, it’s useful for visualizing experimental data, simulation results, and equipment performance curves.”

---

2. Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

3. Basic Line Plot

```python
# Example: Temperature profile in a heat exchanger
temperature = [25, 40, 55, 68, 80]
position = [0, 1, 2, 3, 4]  # meters along the length

plt.plot(position, temperature)
plt.xlabel("Position (m)")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Profile in Heat Exchanger")
plt.show()
```

4. Adding Markers and Styles

```python
plt.plot(position, temperature, marker='o', linestyle='--', color='red')
plt.xlabel("Position (m)")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Profile with Markers")
plt.grid(True)
plt.show()
```

5. Multiple Lines in One Plot

```python
# Pump performance curves
flow = [10, 20, 30, 40, 50]
head_pump1 = [50, 48, 45, 40, 35]
head_pump2 = [55, 53, 50, 45, 40]

plt.plot(flow, head_pump1, label="Pump 1")
plt.plot(flow, head_pump2, label="Pump 2")
plt.xlabel("Flow Rate (m³/h)")
plt.ylabel("Head (m)")
plt.title("Pump Performance Curves")
plt.legend()
plt.grid(True)
plt.show()
```

6. Scatter Plot

```python
# Lab experiment: concentration vs. time
time = [0, 10, 20, 30, 40, 50]
conc = [1.0, 0.82, 0.67, 0.54, 0.43, 0.35]

plt.scatter(time, conc, color='green')
plt.xlabel("Time (min)")
plt.ylabel("Concentration (mol/L)")
plt.title("Batch Reactor Concentration Decay")
plt.show()
```

7. Bar Chart

```python
# Equipment count in plant
equipment = ["Pumps", "Heat Exchangers", "Valves", "Reactors"]
count = [8, 5, 20, 2]

plt.bar(equipment, count, color='skyblue')
plt.xlabel("Equipment Type")
plt.ylabel("Count")
plt.title("Equipment Distribution in Plant")
plt.show()
```

8. Customizing the Plot

```python
plt.plot(position, temperature, color='purple', linewidth=2, marker='o')
plt.xlabel("Position (m)", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.title("Customized Temperature Profile", fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

9. Practical Chemical Engineering Example

```python
# Comparing actual vs. simulated heat exchanger outlet temperatures
length = [0, 1, 2, 3, 4, 5]
actual_T = [25, 38, 50, 62, 74, 80]
simulated_T = [25, 36, 48, 60, 72, 78]

plt.plot(length, actual_T, 'o-', label="Actual Data")
plt.plot(length, simulated_T, 's--', label="Simulation")
plt.xlabel("Length (m)")
plt.ylabel("Temperature (°C)")
plt.title("Heat Exchanger Temperature Profile")
plt.legend()
plt.grid(True)
plt.show()
```
